```python
from openai import OpenAI
import pandas as pd

# Set your OpenAI API key
client = OpenAI(api_key="your-api-key")
```

```python
MODEL="gpt-4o"

def process_text(text):
    try: 
        response = client.chat.completions.create(
            model=MODEL,
            messages=[
                {"role": "system", "content": '''
             You are a privacy law expert. I will first give you a list of standards to grade a privacy policy, and you don't need to reply. "It" in the standards refers to the privacy policy. Next, I will give you a privacy policy. You need to check through the list of standards, and decide whether it's true or false depending on the privacy policy I give to you. If it's true, add one point. Finally, give me the total point of the privacy policy. Your reply wil be formatted as a number.
             
             This is the standards to grade the privacy policy:
             1. It has no statement restricting user rights nor reducing the lgal obligations of the enterprise.
             2. It has a date indicating when was the newest version updated.
             3. It gives at least two definitions of "personal information", "non -personal information", and "personal sensitive information".
             4. It informs users of the purpose of collecting personal information.
             5. It lists all the business functions that use personal information, not using breviary such as "like, for example, such as, etc."
             6. It informs users all the types of personal information being collected.
             7. It informs users the collection methods of the personal information.
             8. The collection of sensitive personal information is listed separatedly, and promise to obtain consent separately. Sensitive personal information includes biometrics, religious beliefs, specific identities, medical health, financial accounts, trajectory trajectory, as well as children under the age of 14.
             9. If biometric information is collected, it should promise not to store the original data.
             10. If the collection method is indirect (sharing, transfer, etc.) , it should inform users of the source where the collected personal information is from, and promise to confirm that the source obtains the personal information legally. 
             11. It informs users the preservation method of personal information, and the time limit for storage. It promises that this is the shortest time necessary. It promises to delete or make data anonymous after the storage time. It specifies what happens if you exceed that time.
             12. It informs users whether personal data is stored domestic or abroad. If oversea transmission is involved, it should list all the personal information being tranfered, and the party who receives.
             13. It informs users the method of personal information anonymization, and whether personal information is used for automated decision-making.
             14. It informs users where the product personalizes user's experience, and provide convenience methods to reject.
             15. It informs users that in special situation (like when requested by the court), personal information will be used without requiring users' consent.
             16. It promises to inform and acquire consent again if the type of personal information collected is changed.
             17. It distinguishes core business functions and additional business functions, and explains in detail
             18. For additional business functions, it promises to obtain users' consent separately, and will not affect core business functions.
             19. It guarantees that users can access, modify or delete the personal information being collected, and explain the operation method and restrictions in detail.
             20. It protects users' rights to withdraw consent, or delete their accounts. 
             21. It promises to delete or anonymize users personal information after consent being withdrawed.
             22. It provides a way for users to transfer their personal information to other products.
             23. It has special rules to deal with the personal information related to the deceased.
             24. It provides the name of company, contact information, and registered address.
             25. It provides effective complaint channel for users to complaint issues regarding personal information.
             26. It informs users any third-party service related in the process of collecting personal information. 
             27. It inform users of the risks that may exist in personal information, and promise to take responsibility if security incidents happens.
             28. It informs users possible consequences if they do not agree with personal information protection policies.
             29. It promises to have procedures to prevent the loss, improper use, leak, damage or tampering of personal information.
             30. It describes what measures to take if a privacy incident happens, like to initiate emergency plans, and notify the affected users timely, report to the relevant authorities of the government.
             31. It lists the rules for children under the age of 14 separately, and promises to obtain the consent of their guardians. 
             32. It promises to take more strict protection measures for personal information of those under age 18. 
             33. It reserves the right to modify the privacy policy.
             34. It informs users of the notification method when the privacy policy is changed.
             35. It informs users of using Cookies, webstream and other tracking technologies.
             36. It states that the company is not responsible for the privacy policy of any third-party, and it's users responsibility to check and determine whether to accept.
             37. It states where users can go for jurisdiction if users have controversy with the company regarding the privacy policy.
             38. It explains whether user data will be used in training AI models.
             39. It refers to privacy laws and regulations in California.
             40. It has links to the previous versions of the privacy policy.
             '''},
             {"role": "user", "content": text}
            ])
        return response.choices[0].message.content.strip()
    except:
        return "too long"
```

```python
# Read the CSV file
df = pd.read_csv('Top100_06_14_Cleaned.csv')
text = df['cleaned_text'][0]
```

```python
# for the first 5 rows in df, process the text
for i in range(1):
    text = df['cleaned_text'][i]
    print(df['Name'][i], ":", process_text(text))
```

```python
# Specify the column to process
column_to_process = 'cleaned_text'

# Process each row in the specified column
df['point'] = df[column_to_process].apply(process_text)

# Write the results to a new CSV file
df.to_csv('output.csv', index=False)

print("Processing complete. Results saved to output.csv.")
```

```python
# save df column Name,Age_rating,Seller,Price,Category,point to a new csv file
df[['Name','Age_rating','Seller','Price','Category','point']].to_csv('points.csv', index=False)
```
