```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('points.csv')
```

```python
# put df['Seller'] to str type
df['Seller'] = df['Seller'].astype(str)
```

```python
# group by Seller, count, and sort in descending order
df.groupby('Seller').size().sort_values(ascending=False)
```

```python
# group by Seller, count, and sort in descending order
df.groupby('Category').size().sort_values(ascending=False)
```

```python
import numpy as np
import ast
df['2D'] = df['2D'].apply(ast.literal_eval)
```

```python
df['2D'].head()
```

```python
# use df['2D'] as coordinates, create a scatter plot
# color by Seller
plt.scatter(df['2D'].apply(lambda x: x[0]), df['2D'].apply(lambda x: x[1]), c=df['Seller'].apply(lambda x: hash(x) % 10000))
# x and y both range from -10 to 10
plt.xlim(-10, 10)
plt.ylim(-10, 10)
# width and height are equal
plt.gca().set_aspect('equal', adjustable='box')
plt.show()
```

```python
# delete all df['point'] == "too long"
df = df[df['point'] != 'too long']
```

```python
df['point'] = df['point'].astype(int)
```

```python
# df sort by point
df = df.sort_values('point')
```

```python
# create a scatter plot from points, x = length, y = point, color by Category
plt.scatter(df['length'], df['point'], c=df['Category'].apply(lambda x: hash(x) % 10000))
# show the color legend
plt.xlim(0, 100000)
plt.colorbar()
plt.show()
```
