```python
import re
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
pd.set_option('display.max_colwidth', None)
```

```python
df = pd.read_csv('Top100_06_14_Cleaned.csv')
```

```python
AI_search_words = [' algorithm ', ' machine ', 'GPT ', ' AI ', ' language model ']
```

```python
Third_party_search_words = [' third party ', ' third-party ', ' thirdparty ', ' third-parties ', ' third parties ']
```

```python
# find sentences that contain any of the search words, save the index into a list
data = []
for i in range(len(df)):
    text = df['cleaned_text'][i]
    sentences = re.split(r'[.!?:\[\]]', text)
    count = len(sentences)
    AI_search_results = []
    Third_party_search_results = []
    for j, sentence in enumerate(sentences):
        for word in AI_search_words:
            if word in sentence:
                AI_search_results.append([j, sentence])
                break
        for word in Third_party_search_words:
            if word in sentence:
                Third_party_search_results.append([j, sentence])
                break
    dic = {'index':i,'count': count, 'AI': AI_search_results,'Thirdparty':Third_party_search_results, 'name': df['Name'][i]}
    data.append(dic)
```
